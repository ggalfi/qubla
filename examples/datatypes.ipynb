```python
import sys
sys.path.insert(0, '../pypkg')

import qubla as qbl
```

```python
myqm = qbl.QuantumLogic()

myqm.compileSource('''
    function reorder(x){    
        local lx = list(x);
        return uword([lx[2], lx[0], lx[3], lx[1]]);
    }
    
    for(i : seq(16)){
        local x = uword{4}(i);
        local y = reorder(x);
        print(str(x) + " + " + str(y) + " = " + str(y + x));
    }
    H = {0 : 1 / sqrt(2), 1 : 1 / sqrt(2)};
    U = {0: {1:exp(1.0i*pi/4)}, 1: {0:exp(-1.0i*pi/4)}};
    print(U);
    arrqb = alloc(4);
    for(i : seq(4))
        arrqb[i] = qstate(H)[0];
    x = word(arrqb);
    print(H);
    print(x);
    y = reorder(x);
    z = x+y;
    applyop(z[[0]], U);
    output(z);
    print(z);
''')

print()
print('####')
print('Quantum Logic Steps:')
for s in myqm.arrstep:
    print(str(s))
    
print()
print('####')
print('QBits:')
for qb in myqm.arrqb:
    print(str(qb))
    
print()
print('####')
print('Inputs:')
for obj in myqm.arrinp:
    print(str(obj))

print()
print('####')
print('Outputs:')
for obj in myqm.arrout:
    print(str(obj))    

print()
print('####')
print("Number of qubits: " + str(myqm.nqb))
```
